```python
def slope(x1, y1, x2, y2):
    if  (x2-x1) == 0:
        return 90
    else: return abs((180/np.pi)*np.arctan((y2-y1)/(x2-x1)))
```

```python
   
def count_stair(image_path):
    
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)


    # Convert the image to gray-scale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    G_blur =cv2.GaussianBlur(gray, (5,5),1)

    edges = cv2.Canny(G_blur, 50, 200)

    lines = cv2.HoughLinesP(edges,rho=1,theta=np.pi/180,threshold=50, minLineLength=0, maxLineGap=250)

    # rho - Distance resolution of the accumulator in pixels.

    # theta - Angle resolution of the accumulator in radians.

    # minLineLength - Minimum length of line. Line segments shorter than this are rejected.

    # maxLineGap - Maximum allowed gap between line segments to treat them as single line.

    # filter vertical lines 

    count=0
    print(lines.shape)
    for line in lines:

        x1, y1, x2, y2 = line[0]

        angle=slope(x1, y1, x2, y2)


        if  angle>2 :
            continue;

    #     print('Coordinates',x1, y1, x2, y2 )
    #     print('slope ',slope(x1, y1, x2, y2))
        count+=1
        cv2.line(img, (x1, y1), (x2, y2), (255, 0, 0), 1)




    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(14,5),)
    ax1.imshow(image)
    ax1.set_title('orignal image')
    # ax2.set_title('GaussianBlur')
    # ax2.set_title('Canny edge')

    ax2.set_title('lined ')


    # ax2.imshow(G_blur)
    # ax2.imshow(edges)
    ax2.imshow(img)


    # plt.imshow(img)
    print('Number of lines ',count)
    print('Number of Stairs  ',count/2)
    
    return count/2
```
